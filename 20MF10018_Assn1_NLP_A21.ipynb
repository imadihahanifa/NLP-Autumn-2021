````python
# Importing modules
import nltk
nltk.download('punkt') # For tokenizers
from nltk.tokenize import word_tokenize,sent_tokenize
````

````python
# *** Write code for preprocessing the corpus ***
# Print first 5 sentences of your preprocessed corpus *** Write code ***
# Print first 5 words/tokens of your preprocessed corpus *** Write code ***


import regex as re

def preprocess(raw_corpus):
    # To convert the text to lower case to make it case insensitive
    txt = raw_corpus.lower()
    
    # To Remove any trailing or leading white space
    txt = txt.strip(" ")

    # To make all white space a single space
    txt = re.sub(pattern = r'\s+', repl = " ", string = txt) 

    # The text contains characters a-z, A-Z, ., ', ? and ,
    txt = re.sub('[^a-zA-Z\.\,\'\?]', ' ', txt)

    # To replace all newlines and blanklines with special strings (" ")
    txt = re.sub(pattern = r'\n', repl = " ", string = txt)
    txt = re.sub(pattern = r'\n\n', repl = " ", string = txt)

    # To extract the word token from the text
    word_tokens = word_tokenize(txt)

    # To extract the sentence tokens from the text
    sentence_tokens = sent_tokenize(txt)

    # Return the words, sentence tokens list
    return word_tokens, sentence_tokens, txt

word_tokens, sentence_tokens, pre_processed_corpus = preprocess(rawReadCorpus)
# Print first 5 sentences of your preprocessed corpus *** Write code ***

print("The first 5 sentences of the preprocessed corpus are: \n")
print( "\n ", sentence_tokens[0])
print("\n ",sentence_tokens[1])
print("\n ", sentence_tokens[2])
print( "\n ", sentence_tokens[3])
print("\n ", sentence_tokens[4])


# Print first 5 words/tokens of your preprocessed corpus *** Write code ***
print("\nThe first 5 words of the preprocessed corpus are as follows:")
print(" ", word_tokens[0])
print(" ", word_tokens[1])
print(" ", word_tokens[2])
print(" ", word_tokens[3])
print(" ", word_tokens[4])



````

````python
from nltk.util import ngrams
unigrams=[]
bigrams=[]
trigrams=[]

for content in sentence_tokens: # *** Write code ***
    words = word_tokenize(content)

    unigrams.extend(words)
    bigrams.extend(ngrams(words,2))
    ##similar for trigrams 
    # *** Write code ***
    trigrams.extend(ngrams(words,3))

print ("Sample of n-grams:\n" + "-------------------------")
print ("--> UNIGRAMS: \n" + str(unigrams[:5]) + " ...\n")
print ("--> BIGRAMS: \n" + str(bigrams[:5]) + " ...\n")
print ("--> TRIGRAMS: \n" + str(trigrams[:5]) + " ...\n")

# list of unigram, bigram & trigram after removing those that 
# totally contain only articles, prepositions, determiners
# Eg. For bigrams, don't remove items like ("a", "boy") --> where not all are 
#     articles, prepositions, determiners
#     But remove items like ("in", "the") --> where all are articles, prepositions, determiners
# Similarly, for unigrams and trigrams

def ngram_filter(ngram_list,n):
  stopwords_set = set(stopwords.words('english'))
  filtered_ngram_list = []

  if n == 1:
    for ngram in ngram_list:
      stopword_count_in_ngram = 0

      # in unigram ngram and gram are the same since it contains only one word
      if ngram in stopwords_set:
        stopword_count_in_ngram += 1

      if stopword_count_in_ngram != n:
        filtered_ngram_list.append(ngram)
  else:
    for ngram in ngram_list:
      stopword_count_in_ngram = 0

      for gram in ngram:
        if gram in stopwords_set:
          stopword_count_in_ngram += 1

      if stopword_count_in_ngram != n:
        filtered_ngram_list.append(ngram)

  return filtered_ngram_list

unigrams_Processed = ngram_filter(unigrams, 1)
bigrams_Processed = ngram_filter(bigrams, 2)
trigrams_Processed = ngram_filter(trigrams, 3)

print ("Sample of n-grams after processing:\n" + "-------------------------")
print ("--> UNIGRAMS: \n" + str(unigrams_Processed[:5]) + " ...\n")
print ("--> BIGRAMS: \n" + str(bigrams_Processed[:5]) + " ...\n")
print ("--> TRIGRAMS: \n" + str(trigrams_Processed[:5]) + " ...\n")

def get_ngrams_freqDist(n, ngramList):
    #This function computes the frequency corresponding to each ngram in ngramList 
    #Here, n=1 for unigram, n=2 for bigram, etc.
    #ngramList = list of unigrams when n=1, ngramList = list of bigrams when n=2
    #Returns: ngram_freq_dict (a Python dictionary where key = a ngram, value = its frequency)
    
    # *** Write code ***
    ngram_freq_dict = {}

    for ngram in ngramList:
      if ngram not in ngram_freq_dict.keys():
        ngram_freq_dict[ngram] = 1
      else:
        ngram_freq_dict[ngram] += 1
    
    return ngram_freq_dict

unigrams_freqDist = get_ngrams_freqDist(1, unigrams)
unigrams_Processed_freqDist = get_ngrams_freqDist(1, unigrams_Processed)
bigrams_freqDist = get_ngrams_freqDist(2, bigrams_Processed)
bigrams_Processed_freqDist = get_ngrams_freqDist(2, bigrams_Processed)
trigrams_freqDist = get_ngrams_freqDist(3, trigrams_Processed)
trigrams_Processed_freqDist = get_ngrams_freqDist(3, trigrams_Processed)         

# Print function
def print_top_10(text,freqDist):

  print('*'*50)
  sorted_dictionary = sorted(freqDist.items(), key=lambda x: x[1], reverse=True)

  index = 1
  print(text)
  for i in sorted_dictionary:
    print("Rank " + str(index)  + " => ", i[0])
    index +=1
    if (index > 10):
      break

  print('*'*50)

# Print top 10 unigrams, having highest frequency as in unigrams_freqDist
# *** Write code ***
print_top_10("Unigrams", unigrams_freqDist)

# Print top 10 unigrams, having highest frequency as in unigrams_Processed_freqDist
# *** Write code ***
print_top_10("Preprocessed unigrams", unigrams_Processed_freqDist)

# Print top 10 bigrams, having highest frequency as in bigrams_freqDist
# *** Write code ***
print_top_10("Bigrams", bigrams_freqDist)

# Print top 10 bigrams, having highest frequency as in bigrams_Processed_freqDist
# *** Write code ***
print_top_10("Preprocessed Bigrams", bigrams_Processed_freqDist)

# Print top 10 trigrams, having highest frequency as in trigrams_freqDist
# *** Write code ***
print_top_10("Trigrams", trigrams_freqDist)

# Print top 10 trigrams, having highest frequency as in trigrams_Processed_freqDist
# *** Write code ***
print_top_10("Preprocessed Trigrams", trigrams_Processed_freqDist)
````

## **Next three words' Prediction using Smoothed Models**

For a bigram model, add-one smoothing is defined by $P_{Add-1}(w_i|w_{i-1})=\frac{count(w_{i-1},w_i)+1}{count(w_{i-1})+V}$.
That is, pretend we saw each word one more time than we did.

You have two tasks here.

First, compute the smoothed bigram and trigram models from the bigrams_freqDist and trigrams_freqDist you calculated above (use the unprocessed models). Second, using these smoothed models, predict the next 3 possible word sequences for testSent1, testSent2 and testSent3, using your smoothed models.

As for example, for the string 'Raj has a' the answers can be as below:

````python

# *** Write code ***
def bigram_smoothed_model(unigrams_freqDist, bigrams_freqDist, sentence, no_of_words):
    
    prob = 0
    prev_word = sentence.split(' ')[-1]
    word = ""
    next_words_list = []
    for i in range(no_of_words):
        
        prob = 0
        for key, value in unigrams_freqDist.items():

            bigram = tuple((prev_word, key))
            
            if bigram in bigrams_freqDist.keys():
                count_bigram = bigrams_freqDist[bigram]
            else:
                count_bigram = 0
            
            if prev_word in unigrams_freqDist.keys():
              count_unigram = unigrams_freqDist[prev_word]
            else:
              count_unigram = 0

            temp_prob = (count_bigram + 1) / (count_unigram + len(unigrams_freqDist.keys()))
            if temp_prob >= prob:
                prob = temp_prob
                word = key

        next_words_list.append(word)
        prev_word = word

    return next_words_list

def trigram_smoothed_model(unigrams_freqDist, bigrams_freqDist, trigrams_freqDist, sentence, no_of_words):

    prob = 0
    prev_word = sentence.split(' ')[-1]
    prev_to_prev_word = sentence.split(' ')[-2]
    word = ""
    next_words_list = []

    for i in range(no_of_words):

        prob = 0

        for key, value in unigrams_freqDist.items():

            trigram = tuple((prev_to_prev_word, prev_word, key))
            bigram = tuple((prev_to_prev_word, prev_word))

            if trigram in trigrams_freqDist.keys():
                count_trigram = trigrams_freqDist[trigram]
            else:
                count_trigram = 0
            
            if bigram in bigrams_freqDist.keys():
                count_bigram = bigrams_freqDist[bigram]
            else:
                count_bigram = 0

            temp_prob = (count_trigram + 1) / (count_bigram + len(unigrams_freqDist.keys()))
            
            if temp_prob >= prob:
                prob = temp_prob
                word = key

        next_words_list.append(word)
        prev_to_prev_word = prev_word
        prev_word = word

    return next_words_list




print()
print("For testSent 1: ")

bigram_words_list = bigram_smoothed_model(unigrams_freqDist, bigrams_freqDist, testSent1, 3)

bigram_predicted_sentence = testSent1 + " "
for word in bigram_words_list:
  bigram_predicted_sentence = bigram_predicted_sentence + word + " "

print("Bigram predicted sentence: ",bigram_predicted_sentence)

trigram_words_list = trigram_smoothed_model(unigrams_freqDist, bigrams_freqDist, trigrams_freqDist, testSent1, 3)

trigram_predicted_sentence = testSent1 + " "
for word in trigram_words_list:
  trigram_predicted_sentence = trigram_predicted_sentence + word + " "

print("Trigram predicted sentence: ",trigram_predicted_sentence)
print()


print()
print("For testSent 2: ")

bigram_next_words_list = bigram_smoothed_model(unigrams_freqDist, bigrams_freqDist, testSent2, 3)

bigram_predicted_sentence2 = testSent2 + " "
for word in bigram_next_words_list:
  bigram_predicted_sentence2 = bigram_predicted_sentence2 + word + " "

print("Bigram predicted sentence: ",bigram_predicted_sentence2)

trigram_next_words_list = trigram_smoothed_model(unigrams_freqDist, bigrams_freqDist, trigrams_freqDist, testSent2, 3)

trigram_predicted_sentence2 = testSent2 + " "
for word in trigram_next_words_list:
  trigram_predicted_sentence2 = trigram_predicted_sentence2 + word + " "

print("Trigram predicted sentence: ",trigram_predicted_sentence2)
print()


bigram_next_words_list = bigram_smoothed_model(unigrams_freqDist, bigrams_freqDist, testSent3, 3)

bigram_predicted_sentence3 = testSent3 + " "
for word in bigram_next_words_list:
  bigram_predicted_sentence3 = bigram_predicted_sentence3 + word + " "

print()
print("For testSent 3: ")
print("Bigram predicted sentence: ",bigram_predicted_sentence3)

trigram_next_words_list = trigram_smoothed_model(unigrams_freqDist, bigrams_freqDist, trigrams_freqDist, testSent3, 3)

trigram_predicted_sentence3 = testSent3 + " "
for word in trigram_next_words_list:
  trigram_predicted_sentence3 = trigram_predicted_sentence3 + word + " "

print("Trigram predicted sentence: ",trigram_predicted_sentence3)
print()


````
